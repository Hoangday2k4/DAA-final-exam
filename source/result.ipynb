```python
import matplotlib.pyplot as plt

# Dữ liệu
grids = ["4_2", "4_3", "4_4", "4_5", "8_5", "8_6", "8_7", "16_5"]
times = [3.49, 2.62, 8.08, 17.45, 38.69, 91.84, 231.64, 92.48]

# Vẽ biểu đồ
plt.figure(figsize=(8, 5))
plt.plot(grids, times, marker='o', linestyle='-', color='b', label="Thời gian chạy (s)")
plt.xlabel("Kích thước lưới và số lượng tác tử")
plt.ylabel("Thời gian chạy trung bình (ms)")
plt.title("Biểu đồ so sánh thời gian chạy thuật toán CBS")
plt.legend()
plt.grid(True)

# Hiển thị biểu đồ
plt.show()
```
